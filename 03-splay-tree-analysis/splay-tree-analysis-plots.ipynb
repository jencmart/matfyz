```python
def crate_range(n, a, start, step):      
    y = []
    x = []
    rang = int(n/100*5+1)

    # proggression 1
    for i  in range(1,rang):
        y.append(a + i) # 75, 76, 77, 78, 79, 80
        x.append(start + i*step) # 50, 46, 4
    return (x,y)

n = 100
a = list(range(0,n))
b = [4,10,60,96,42,55,54,53,52,51,50,26,27,28,29,30,13,33,19,1,94,74,92,83,24,71,11,67,48,38,80,68,20,63,79,36,46,15,78,59,85,97,77,32,47,86,76,23,81,37,87,49,7,8,40,88,66,34,70,35,89,25,6,2,73,90,65,44,5,39,99,22,75,3,82,12,18,9,14,56,45,62,84,0,69,31,21,98,43,64,95,61,16,93,72,41,17,58,57,91  ]

from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 4), dpi=200, facecolor='w', edgecolor='k')
title_size=13
plt.title(r'random permutation $\{x_0, x_1,\dots,x_{99}\}$ with 4 arithmetic progressions', fontsize=title_size)
label_size = 15
plt.xlabel(r'$i$', fontsize=label_size)
plt.ylabel(r'$x_i$', fontsize=label_size)

style = 'rx-'
plt.plot(a,b, "ob-")

x, y = crate_range(n, n/4 , n/10, 1)
plt.plot(x,y, style)

x, y = crate_range(n, n/2 , n/10, -1)
plt.plot(x,y, style)

x, y = crate_range(n, n/4*3 , n/2, -4)
plt.plot(x, y, style)

x, y = crate_range(n, 17*n/20, 2*n/5, 5)
plt.plot(x,y, style)

plt.show()
```
